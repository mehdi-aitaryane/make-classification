```python
from datasets import make_classification, count_occurances
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_classification(n_samples=10000, n_features=2, n_classes=2, noise=0.0, balanced=True, random_state=555)
```

```python
print("X shape is ", X.shape)
print("y shape is ", y.shape)
```

```python
count_occurances(y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.title("Balanced Syntactic Dataset For Binary Classification")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
X, y = make_classification(n_samples=10000, n_features=2, n_classes=2, noise=0.0, balanced=False, random_state=74)
```

```python
print("X shape is ", X.shape)
print("y shape is ", y.shape)
```

```python
count_occurances(y)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
plt.title("Unbalanced Syntactic Dataset For Binary Classification")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
X, y = make_classification(n_samples=986, n_features=3, n_classes=4, noise=0.0, balanced=True, random_state=111)
```

```python
print("X shape is ", X.shape)
print("y shape is ", y.shape)
```

```python
count_occurances(y)
```

```python
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(X[:, 0], X[:, 1], X[:, 2], c = y)
plt.title("Balanced Syntactic Dataset For Multi Classification")
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')

# Rotate the plot (change the angles as needed)
ax.view_init(elev=30 , azim=-78.1 )
plt.show()
```
